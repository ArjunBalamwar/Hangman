```python
import random

countries = [
    "France",
    "United-States",
    "China",
    "Spain",
    "Italy",
    "Turkey",
    "United-Kingdom",
    "Germany",
    "Russia",
    "Malaysia",
    "India"
]

def play_again():
    print("Want to play again?(y/n)")
    reply = input().lower()
    if reply == 'y':
        play_game()
    elif reply == 'n':
        pass
    else:
        print("Sorry didn't understand")
        play_again()


def random_country():
    country = random.choice(countries).upper()
    return country

def play_game():
    country = random_country()

    display = " ".join("_" * len(country)) + " "
    country = " ".join(country) + " "
    life = 7
    
    while "_" in display and life > 0:
        incorrect_guess = True
        if life >1:
            print("Lives left: "+ str(life))
        else:
            print("Life left: "+ str(life))
        print(display)
        print("Enter word/letter: ")
        guess = " ".join(input().upper())
        size = len(guess)

        if len(country) == size and guess == country:
            display = country
            print("You won")
            break
        else:
            for index in range(len(country)-size):
                if guess in country[index:size+index]:
                    display = display[:index] + guess + display[size+index:]
                    incorrect_guess = False
                    
        if incorrect_guess:
            life = life - 1
    
    if "_" in display:
        print("You lost")
    else:
        print("You Won!")
    
    play_again()

print("USE HYPHEN FOR SPACE ('-')")
play_game()

```
